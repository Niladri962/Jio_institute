```python
# Restaurant Ordering Program

print("Welcome to ABC Restaurant")

# Seat location
seat = input("Where would you like to sit? (Indoor / Outdoor): ")

if seat.lower() == "indoor":
    print("You selected Indoor seating.")
else:
    print("You selected Outdoor seating.")

# Number of guests
guests = int(input("Enter number of guests: "))

if guests <= 2:
    print("Small table reserved.")
elif guests <= 5:
    print("Medium table reserved.")
else:
    print("Large table reserved.")

# Veg / Non-Veg menu
menu = input("Choose menu type (Veg / NonVeg): ")

if menu.lower() == "veg":
    print("Veg Menu:")
    print("1. Paneer Butter Masala")
    print("2. Veg Biryani")
    print("3. Dal Tadka")
else:
    print("Non-Veg Menu:")
    print("1. Chicken Curry")
    print("2. Mutton Biryani")
    print("3. Fish Fry")

# Payment mode
payment = input("Choose payment mode (Cash / Card / UPI): ")

if payment.lower() == "cash":
    print("You selected Cash payment.")
elif payment.lower() == "card":
    print("You selected Card payment.")
elif payment.lower() == "upi":
    print("You selected UPI payment.")
else:
    print("Invalid payment option.")

print("Thank you for visiting ABC Restaurant!")
```
