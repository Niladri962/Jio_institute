A set is:

Unordered

Mutable (can be changed)

Does not allow duplicate values

Sets are written using curly braces { }

```python
numbers = {1, 2, 3, 4}
print(numbers)
```

```python
numbers.add(5)
print(numbers)
```

```python
numbers.remove(2)
print(numbers)
```

```python
numbers.discard(4)
print(numbers)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
```

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
```
