a Tuple is a collection data type used to store multiple items in a single variable.

A tuple is ordered but immutable, which means its elements cannot be changed after creation.

Tuples are written using parentheses ( )

```python
numbers = (10, 20, 30, 40)
print(numbers)
```

```python
numbers = (10, 20, 30, 40)
print(numbers[1])
```

```python
numbers = (1, 2, 3, 2, 4, 2)
print(numbers.count(2))
```
