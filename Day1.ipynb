DataType

1. Numeric: Integer, Float, Complex
2. Sequence: String, List, Tuple
3. Dictionary
4. Set
5. Boolean
6. Binary: Bytes, Bytearray, Memoryview

```python
# Int data type
a = 10
print("Value of a:", a)
print("Data type of a:", type(a))
```

```python
#Float data type
a=5.6
print(a)
print(type(a))
```

```python
# Complex data type
c = 3 + 4j
print("Value of c:", c)
print("Data type of c:", type(c))
```

```python
# String Data Type
my_name = "Niladri"
print(my_name)
print(type(my_name))
```
