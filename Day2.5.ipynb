**DICTIONARY**
1. Key:Value Pairs
2. Unordered
3. Mutable
4. Doesn't allow duplicate

```python
student = {
    "name": "Niladri",
    "age": 23,
    "course": "B.Tech"
}

print(student)
```

```python
print(student.keys())
```

```python
print(student.values())
```

```python
dir(dict)
```

```python
student = {
    "name": "Niladri",
    "age": 23,
    "course": "B.Tech"
}

print(student.update({'age':'21'}))
print(student)
```

```python
student = {
    "name": "Niladri",
    "age": 23,
    "course": "B.Tech"
}

print(student.get('course'))
```
